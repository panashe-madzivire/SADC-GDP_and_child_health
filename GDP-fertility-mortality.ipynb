```python
#import packages
import pandas as pd
import glob
import pymysql
```

Add the excel sheets

```python
#save all files to bused in this analysis to a list using glob
all_excel = [i for i in glob.glob('*.xlsx')]
all_excel
```

```python
#create a list to store each dataframe separately

all_dataframes_store = []
for file in all_excel:
    df = pd.read_excel(file)
    all_dataframes_store.append(df)
    
all_dataframes_store[0].head()
```

```python
all_dataframes = all_dataframes_store.copy()
```

```python
#create a list to reverse the years
col =all_dataframes[0].columns[3:]
col=col[::-1]
col=col.insert(0,'Country')
col
```

```python
# clean the tables
n=[0,3]

for num in n:
    all_dataframes[num]=all_dataframes[num].drop([0,1, 18, 19, 20, 21, 22, 23])
    all_dataframes[num]=all_dataframes[num].drop(['Unnamed: 1', 'Unnamed: 2'], axis = 1)

    all_dataframes[num].rename(columns={"Time period":"Country"}, inplace=True)
        
    all_dataframes[num] = all_dataframes[num][col]
    
    all_dataframes[num].loc[15, ['Country']]= 'Tanzania'
    
all_dataframes[0]
```

```python
n = [1,2]
col = col.insert(-1, '2020')

for num in n:
    all_dataframes[num]=all_dataframes[num].drop(list(range(17,22)))
    all_dataframes[num]=all_dataframes[num].drop(['Series Name', 'Series Code', 'Country Code'], axis = 1)

    all_dataframes[num].columns = col
    
    all_dataframes[num].loc[3, ['Country']]= 'Democratic Republic of the Congo'
    
all_dataframes[2]
```

```python
for i, df in enumerate(all_dataframes):
    all_dataframes[i] = all_dataframes[i].melt(id_vars=['Country'], var_name= 'Year',
                               value_name = f'{all_excel[i][:-5]}')
    
all_dataframes[0].head()
```

```python
#create a function that I can enter a dataframe into to make it an exportable data table
def CreateTable(df,SQL_table):
    insert_query = f'INSERT INTO {SQL_table} VALUES'
    values_array = df.values
    df_values = values_array.tolist()
    df_values_export =[]

    for row in values_array:
        df_values_export.append(tuple(row))
    insert_query +=str(df_values_export)[1:-1] + ';'
    

    return insert_query
```

```python
SADC_fertility_rate = CreateTable(all_dataframes[0], 'SADC_fertility_rate')
SADC_GDP_growth_per_capita = CreateTable(all_dataframes[1], 'SADC_GDP_growth_per_capita')
SADC_GDP_per_capita = CreateTable(all_dataframes[2], 'SADC_GDP_per_capita')
Under_5_mortality = CreateTable(all_dataframes[3], 'Under_5_mortality')
```
